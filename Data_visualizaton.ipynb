```python

crimes_counts = df.groupby(['Hood_ID', 'MCI'])['event_unique_id'].count()
crimes_counts_df = crimes_counts.to_frame().unstack('MCI')
```

```python
crimes_counts_df
```

```python
incomes_by_neighborhood = df[['Hood_ID','income']].groupby('Hood_ID').mean()
incomes_by_neighborhood.head()
```

```python
crimes_with_income = pd.merge(crimes_counts_df,incomes_by_neighborhood, on='Hood_ID')
crimes_with_income.columns = ['Assault','Auto Theft','Break and Enter','Robbery','Theft Over','Average Income']
crimes_with_income.head()
```

```python
MCIs = list(df.groupby('MCI').groups)
MCIs
```

```python
plt.clf()
for MCI in MCIs:
    scatter = plt.scatter(crimes_with_income['Average Income'],crimes_with_income[MCI], s=10, label = MCI) 
plt.legend()    

plt.xlim(20000,140000)
plt.ylim(0,3000)
plt.title('Crime Types vs Average Income per Neighborhoud')
plt.xlabel('Average yearly gross income (CAD)')
plt.ylabel('Crime count')
#plt.legend()
plt.show()
```

```python
from scipy import stats
```

```python
x = crimes_with_income['Average Income']
for MCI in MCIs:
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, crimes_with_income[MCI])
    plt.plot(x,intercept + slope*x,label = MCI)   
plt.legend()
```

```python
education_by_neighborhood = df[['Hood_ID','Postsecondary rate']].groupby('Hood_ID').mean()
education_by_neighborhood.head()
```

```python
crimes_with_education = pd.merge(crimes_counts_df,education_by_neighborhood, on='Hood_ID')
crimes_with_education.columns = ['Assault','Auto Theft','Break and Enter','Robbery','Theft Over','Postsecondary rate']
crimes_with_education.head()
```

```python
plt.clf()
for MCI in MCIs:
    scatter = plt.scatter(crimes_with_education['Postsecondary rate']*1000,crimes_with_education[MCI], s=10, label = MCI) 
plt.legend()    

plt.xlim(10,1000)
plt.ylim(0,3000)
plt.title('Crime Types vs Average education per Neighborhoud')
plt.xlabel('Postsecondary rate*1000')
plt.ylabel('Crime count')
#plt.legend()
plt.show()
```

```python
x = crimes_with_education['Postsecondary rate']*1000
for MCI in MCIs:
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, crimes_with_education[MCI])
    plt.plot(x,intercept + slope*x,label = MCI)   
plt.legend()
```

```python
housing_by_neighborhood = df[['Hood_ID','Unsuitable rate']].groupby('Hood_ID').mean()
housing_by_neighborhood.head()
```

```python
x = crimes_with_housing['Unsuitable rate']*10
for MCI in MCIs:
    slope, intercept, r_value, p_value, std_err = stats.linregress(x, crimes_with_housing[MCI])
    plt.plot(x,intercept + slope*x,label = MCI)   
plt.legend()
```
